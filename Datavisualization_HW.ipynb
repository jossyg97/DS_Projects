```python
percent_missing = df.isnull().sum() * 100 / len(df)
missing_value_df = pd.DataFrame({'column_name': df.columns,
                                 'percent_missing': percent_missing})

percent_missing
```

```python
df["Genre"].value_counts()
```

```python
char_explode = [0.3] * len(df.Genre.value_counts())
df["Genre"].value_counts().plot(kind = 'pie', autopct='%1.0f%%', radius = 2, explode = char_explode)
```
