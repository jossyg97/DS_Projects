### Change date_added to datetime 

```python
df['date_added'] = pd.to_datetime(df['date_added'])
df
```

```python
df.dtypes.value_counts()
```

#### Add a new column for added_year


```python
#create new column called added_year which shows the difference between years
df['added_year'] = pd.DatetimeIndex(df['date_added']).year
```

```python
#change release_year from string to int
df['release_year'] = df["release_year"].astype(str).astype(int)
```

```python
df.head()
```

```python
pd.api.types.infer_dtype(df['added_year'])
```

```python
df.dtypes.value_counts()
```

```python
#find the years difference from year added to netflix and release year to make it into a new column
df['years_difference']= df['added_year'].sub(df['release_year'], axis = 0)
```

```python
df.head()
```

### This is the Movie Dataset


```python
#Create a new dataset that is for all that are movies
group_1 = df.groupby(df.type)
df_new = group_1.get_group("Movie")
df_new
```

### Change Duration from String to Int

```python
#Created a function that checks all that are not nan and changed it to int 
def convert_to_min(row):
    if row['duration'] is not np.nan:
        return int(row['duration'].removesuffix('min'))
    
#changed the duration column from new dataframe to new data which is int
df_new['duration'] = df_new.apply(convert_to_min, axis = 1)
```

Sorted by release_year

```python
df_new.sort_values("release_year")
```

### Check for percentage values of each row that are null 

```python
percent_missing = df_new.isnull().sum() * 100 / len(df_new)
missing_value_df = pd.DataFrame({'column_name': df_new.columns,
                                 'percent_missing': percent_missing})

percent_missing
```

### Check for what movies have a specific genre

```python
def movie_type():
    lst_type= []
    #In this case I am checking for all movies that are documentaries, but can be changed to something else
    mask = df_new['listed_in'].str.contains(r'Comedy', na=True) 
    lst_type.append(df_new.loc[mask, 'title'])
    return lst_type
```

```python
movie_type()
```

### Visualize duration of movie based on release year

```python
plt.scatter(df_new["release_year"], df_new["duration"])
plt.show
```

### Take the average of all columns that are nums based on release year

```python
df_duration_by_year = df_new.groupby(df['release_year']).mean().astype(int)
df_duration_by_year
```

### Visualize the average movie duration based on year

```python
plt.bar(df_duration_by_year['release_year'], df_duration_by_year['duration'], color ='blue',
        width = 0.5)
plt.xlabel("Year")
plt.ylabel("Average Duration in Min")
plt.title("Average Movie's Duration Based on Year")
plt.show()
```
